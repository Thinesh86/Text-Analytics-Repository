```python
url = 'https://shopee.com.my/PUMA-Unisex-Aviator-Running-Shoes-i.341487315.16930565081?sp_atk=c1043022-11bf-479c-8cbb-6b1a052e6450&xptdk=c1043022-11bf-479c-8cbb-6b1a052e6450'
```

```python
url_match = re.search(r'i\.(\d+)\.(\d+)', url)
if url_match:
    shop_id, item_id = url_match.group(1), url_match.group(2)
else:
    print("Invalid URL")
    exit()
```

```python
# Define 5 pages to extract
num_pages = 5
```

```python
# Create and open a CSV file to store the reviews
with open('Shoppe_Reviews.csv', mode='w', newline='', encoding='utf-8') as reviews_file:
    csv_writer = csv.writer(reviews_file)
    csv_writer.writerow(['Reviewer Name', 'Review Date', 'Review Content'])

    # Iterate over the specified number of pages
    for current_page_num in range(num_pages):
        offset_for_page = current_page_num * 20  

        # Construct the ratings URL for the current page
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset_for_page}&shopid={shop_id}&type=0'

        # Fetch ratings data from the Shopee API
        ratings_data = requests.get(ratings_url).json()

        # Extract and write reviewer name, review date, and review content to CSV file
        for rating in ratings_data['data']['ratings']:
            reviewer_name = rating['author_username']
            review_timestamp = rating['ctime']
            review_date = datetime.fromtimestamp(review_timestamp).strftime('%Y-%m-%d %H:%M:%S')
            review_content = rating['comment']
            
            csv_writer.writerow([reviewer_name, review_date, review_content])
            print(f"Reviewer Name: {reviewer_name}\nReview Date: {review_date}\nReview Content: {review_content}\n")
```
